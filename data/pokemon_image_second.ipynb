```python
from splinter import Browser
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
import requests
import shutil

poke_df = pd.read_csv("pokemon_info.csv")
```

```python
poke_list = poke_df['name'].to_list()
poke_names = []
for poke in poke_list:
    pokes = poke.lower()
    poke_names.append(pokes)
```

```python
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
url = f"https://www.pokemon.com/us/pokedex/charizard"
browser.visit(url)


html = browser.html
soup = BeautifulSoup(html, "html.parser")
results = soup.find('div', class_="profile-images")
img = results.find('img', class_="active")['src']
```

```python
poke_pics = []
edge = []

for poke in poke_names:
    try:
        url = f"https://www.pokemon.com/us/pokedex/{poke}"
        browser.visit(url)

        html = browser.html
        soup = BeautifulSoup(html, "html.parser")
        results = soup.find('div', class_="profile-images")
        img = results.find('img', class_="active")['src']
        poke_pics.append(img)
        
    except:
        poke_pics.append(f"Edge for {poke}")
        print(f"Edgecase for {poke}")
        edge.append(poke)
```

```python
for x in range(len(poke_pics)):
    try: 
        r = requests.get(poke_pics[x], stream = True)
        if r.status_code == 200:
            r.raw.decode_content == True
    
            with open(f'p_images/{poke_names[x]}.jpg','wb') as f:
                shutil.copyfileobj(r.raw, f)
            print('Image download successful')
        else:
            print('Unsuccessful')
    except:
        pass
```
