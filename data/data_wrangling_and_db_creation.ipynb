```python
# Import dependencies
import pandas as pd
%store -r poke_entries
```

```python
# Data source: https://www.kaggle.com/rounakbanik/pokemon

pokemon_df = pd.read_csv("csv/pokemon.csv")
pokemon_df.head()
```

```python
# Drop the columns we don't care about

pokemon_ordered = pokemon_df.drop(columns={"base_egg_steps", "base_happiness", "base_total", "capture_rate", 
                                           "experience_growth", "japanese_name"})
pokemon_ordered.head()
```

```python
# Set index to pokedex number
pokemon_ordered = pokemon_ordered.set_index("pokedex_number")
pokemon_ordered.head()
```

```python
# Make dataframe just for pokemon info
pokemon_info = pokemon_ordered[["name", "generation", "classfication", "type1", "type2", "hp", "height_m",
                                "weight_kg", "speed", "percentage_male", "is_legendary"]]
pokemon_info['entry'] = poke_entries
pokemon_info.head()
pokemon_info.to_csv("csv/final_pokemon_info.csv")
```

```python
pokemon_info.count()
```

```python
# Replace NaN values in type 2 column with "none"

pokemon_info["type2"] = pokemon_info["type2"].fillna("none")

pokemon_info.head()
```

```python
# Make datafram for battle info
pokemon_battle_info = pokemon_ordered[["attack", "defense", "sp_attack", "sp_defense", "abilities", "against_bug", 
                                       "against_dark", "against_dragon", "against_electric", "against_fairy", 
                                       "against_fight", "against_fire", "against_flying", "against_ghost", 
                                       "against_grass", "against_ground", "against_ice", "against_normal", 
                                       "against_poison", "against_psychic", "against_rock", "against_steel", 
                                       "against_water"]]
pokemon_battle_info.head()
```

## Connect to Mongo DB to insert data

```python
# Import dependencies
import pymongo
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collections
db = client.pokemon
info_collection = db.info
battle_collection = db.battle
```

```python
# In case collection is already populated, drop it first
delete_col = info_collection.delete_many({})
print(delete_col.deleted_count, " documents deleted.")
```

```python
# Loop through info df to insert into mongodb

for index, row in pokemon_info.iterrows():
    print("Adding pokemon " + str(index), row['name'] + " to the database.")
    
    add_pokemon = {
        '_id': index,
        'name': row['name'],
        'generation': row['generation'], 
        'classfication': row['classfication'], 
        'type1': row['type1'], 
        'type2': row['type2'], 
        'hp': row['hp'], 
        'height_m': row['height_m'],
        'weight_kg': row['weight_kg'], 
        'speed': row['speed'], 
        'percentage_male': row['percentage_male'], 
        'is_legendary': row['is_legendary'],
        'entry': row['entry']
    }

    print (add_pokemon)
    print("\n")
    info_collection.insert_one(add_pokemon)
```

```python
# Check data in db

y = db.info.find_one({"name": "Dragonite"})

print(y)
```

```python
# In case collection is already populated, drop it first
delete_col = battle_collection.delete_many({})
print(delete_col.deleted_count, " documents deleted.")
```

```python
# Loop through info df to insert into mongodb

for index, row in pokemon_battle_info.iterrows():
    print("Adding pokemon #" + str(index) + " to the database.")
    
    # Turn abilities string into a list
    abilities = row["abilities"].strip("[]''").split("', '")
    
    add_pokemon = {
        '_id': index,
        "attack": row["attack"], 
        "defense": row["defense"], 
        "sp_attack": row["sp_attack"], 
        "sp_defense": row["sp_defense"], 
        "abilities": abilities, 
        "against_bug": row["against_bug"], 
        "against_dark": row["against_dark"], 
        "against_dragon": row["against_dragon"], 
        "against_electric": row["against_electric"], 
        "against_fairy": row["against_fairy"], 
        "against_fight": row["against_fight"], 
        "against_fire": row["against_fire"], 
        "against_flying": row["against_flying"], 
        "against_ghost": row["against_ghost"], 
        "against_grass": row["against_grass"], 
        "against_ground": row["against_ground"], 
        "against_ice": row["against_ice"], 
        "against_normal": row["against_normal"], 
        "against_poison": row["against_poison"], 
        "against_psychic": row["against_psychic"], 
        "against_rock": row["against_rock"], 
        "against_steel": row["against_steel"], 
        "against_water": row["against_water"]
    }

    print (add_pokemon)
    print("\n")
    battle_collection.insert_one(add_pokemon)
```
